```python
branch1_stock = [
    [10, 20, 30],
    [40, 50, 60]
]

branch2_stock = [
    [1, 2, 3],
    [4, 5, 6]
]

total_stock = [
    [0, 0, 0],
    [0, 0, 0]
]

for i in range(len(branch1_stock)):
    for j in range(len(branch1_stock[i])):
        total_stock[i][j] = branch1_stock[i][j] + branch2_stock[i][j]

print(total_stock)
```

```python
class Penguin:
  family = "Aptenodytes forsteri"

  def __init__(self, name, age, color):
    self.name = name
    self.age = age

  def swim(self):
    return f'{self.name} can swim.'

  def sing(self, can_sing=False):
    if can_sing:
      return f'{self.name} can sing'
    else:
      return f'{self.name} can NOT sing'

  def dance(self, can_dance=False):
    if can_dance:
      return f'{self.name} can dance'
    else:
      return f'{self.name} can NOT dance'
```

```python
Emperor_chick = Penguin('Emperor Chick Penguin', 2, "Grey" )
```

```python
print(Emperor_chick.swim())
```

```python
print(Emperor_chick.sing(False))
```
